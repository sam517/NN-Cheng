```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
data=[]
with open("abalone_train_5.2") as tsv:
    
    reader=csv.reader(tsv,delimiter=' ', quoting=csv.QUOTE_NONNUMERIC)
    for row in reader:
        data.append(row)

my_array=np.array(data,dtype=np.float64);
my_array.shape
X=my_array[:,1:11]
X.shape
y=my_array[:,0]
y=y.astype(np.uint8)
print (y.dtype, y.shape)
```

```python
#Normalize X

for j in range(X.shape[1]):
    print(j)
    base=min(X[:,j])
    spread=max(X[:,j])-base
    z=X[:,j]
    z=2*(z-base)/spread - 1
    X[:,j]=z
```

```python
#Start counting in 0 for the categories of y
#Here we create a matrix Y that holds 0 and 1's according to 
#the paper for ordinal regression

n=y.shape[0]
nclasses=max(y)-min(y)+1

Y=np.zeros((n,nclasses))

for i in range(n):
    #print(i)
    c=y[i]+1
    for j in range(c):
        #print(j)
        Y[i,j]=1
```

```python
#Placeholders
X_=tf.placeholder(tf.float32, shape=X.shape)
Y_=tf.placeholder(tf.float32, shape=Y.shape)
```

```python
#Model definition

#Hidden units
HU=8

#First layer
W1=tf.Variable(tf.random_uniform([X.shape[1],HU],-1.0,1.0)) #Weights matrix
b1=tf.Variable(tf.zeros([HU])) #Bias
O=tf.nn.tanh(tf.matmul(X_,W1)+b1) #Output from the first layer

#Second layer
W2=tf.Variable(tf.random_uniform([HU,nclasses],-1.0,1.0))
b2=tf.Variable(tf.zeros([nclasses]))
YHat=tf.nn.sigmoid(tf.matmul(O,W2)+b2)
```

```python
#Loss and optimizer definition
#Cross entropy loss function

cost=tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(labels=Y_,logits=YHat))
#Hace lo mismo que la fn de arriba cost=tf.reduce_sum(Y_*-tf.log(tf.sigmoid(YHat))+(1-Y_)*-tf.log(1-tf.sigmoid(YHat)),reduction_indices=[0,1])
#Optimizing the function cost by gradient descent with learning step 0.01
train_step=tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
#Create the session and initialize variables
sess=tf.Session()

sess.run(tf.global_variables_initializer())
```

```python
#Training 

Epochs=2000

error=np.zeros((Epochs))

for e in range(Epochs):
    sess.run(train_step, feed_dict={X_:X,Y_:Y})	#Optimizer step
    print("Epoch=",e)
    error[e]=sess.run(cost,feed_dict={X_:X,Y_:Y})
    print("Cost=",error[e])
```

```python
predictions=sess.run(YHat,feed_dict={X_:X,Y_:Y})

predictions=np.round(predictions,1)

print ("Output:", predictions)
```

```python
Y_Prediction=np.zeros(n)

for i in range(n):
    for j in range(predictions.shape[1]):
        
        if predictions[i,j] > 0.5:
            Y_Prediction[i]=Y_Prediction[i]+1
        else:
            break
```

```python
yc=y+1
correct_prediction=abs(yc-Y_Prediction)<0.5
cast=tf.cast(correct_prediction,"float")
accuracy=tf.reduce_mean(cast)
acc=sess.run(accuracy)
print ("Accuracy",acc)
```
